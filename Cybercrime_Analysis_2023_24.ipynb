```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel data
df = pd.read_excel("cybercrime_data_template.xlsx")

# Create helper column
df["Label"] = df["Victim Type"] + " – " + df["Attack Type"]

# Sort by report volume
df = df.sort_values("Total Reports (Est)", ascending=False)

# Create chart
plt.figure(figsize=(14, 6))
bars = plt.barh(df["Label"], df["Total Reports (Est)"], color="skyblue")

# Add report count labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 100, bar.get_y() + bar.get_height()/2, f"{int(width)}", va='center')

plt.xlabel("Estimated Total Reports")
plt.title("Cybercrime Report Volume by Attack Type (2023–24)")
plt.tight_layout()
plt.gca().invert_yaxis()
plt.show()
```
